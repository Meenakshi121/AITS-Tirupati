### Types of plots

```python
## Line -> is a graph ,shows frequency of data in a number line

x=[10,12,14,16]
y=[4,5,6,7]
plt.plot(x,y,label="Line")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("Line graph")
plt.legend(loc="lower right")
```

```python
import numpy as np
a=np.linspace(0,5,20)
a
```

```python
plt.plot(a,a**2,color='red',label='Line1')
plt.plot(a,a**3,color='green',label='Line2')
plt.plot(a,a**4,color='yellow',label='Line3')
plt.legend()
```

### Bar chart

* along with pyplot,we can use bar() to draw bar chart
* plt.bar()

```python
year=[2020,2005,2023,2000]
goldprices=[4500,3500,5000,3000]
silver=[1500,1100,1800,1000]
plt.bar(year,goldprices,width=0.89,color='green')
plt.bar(year,silver,width=0.89,color='red')
```

```python
x=[1,2,3,4,5,6]
y=[10,11,12,13,14,15]
plt.bar(x,y,color=['c','y','m','r','g','b'],edgecolor='yellow')
```

```python
# horizontal barchart

# plt.barh()

plt.barh(x,y)
```

### Scatter plot

```python
plt.scatter([12,14,16],[16,17,19])
```

```python
x=np.random.randint(100,1000,100)
y=np.random.randint(200,1000,100)
plt.scatter(x,y)
```

### Boxplot

* It gives the summary of numercial data through quartiles
* min
* Inter quartile Range(IQR)
    * lower quartile(25%)
    * Median(50%)
    * Upper Quartile range(75%)
* Max
* plt.boxplot()

```python
number=np.random.randint(0,1000,1000)
plt.boxplot(number)
```

```python
# notched boxplot

plt.boxplot(number,notch=True)
```

### Histogram

* It is a diagram which shows the frequency distribution of numerical data
* plt.hist()

```python
x=[10,90,87,67,56,43]
plt.hist(x)
```

### Piechart()
* plt.pie()

```python
games=['ludo','carroms','chess','football']
scores=[87,67,88,76]
plt.pie(scores,labels=games,autopct='%1.2f%%',explode=[0,1,0,0])
plt.savefig('piechart.png')
```

### Categoical plot

* By default it returns scatter plot
* sns.catplot()

```python
sns.catplot(x="species",y="sepal_length",data=ir)
```

### Categorical distribution of data
* boxplot

```python
sns.catplot(x="species",y="sepal_length",data=ir,kind="box")
```

```python
sns.catplot(data=ir,kind="box")
```

```python
sns.catplot(data=ir,kind="box",orient="h")
```

```python
sns.catplot(x="species",y="sepal_length",data=ir,kind="violin")
```

```python
sns.catplot(x="species",y="sepal_length",data=ir,kind="bar")
```
