### Data analysis

* Data :data is facts and statistics collected together for reference or analysis
* Data analysis:
    * is a process of obtaining raw data and subsequently converting it into information,useful for decision making
    * is a technique to collect , transform and organize data to make future predictions
    * all sorts of data analysis,Analyzing the numerical data with numpy,tabular data with pandas 
    and visualization by matplotlib and exploratory data analysis

### Data Collection

* Primary data: 
    * refers to first hand data generated by the research himself/herself
    * real time data
    * Surveys, Observations,experiments,questionanaire,personal interviews etc
    * Expensive and takes long time
    
* Secondary data:
    * data which is collected by someone else earlier
    * Past data
    * Websites, books, journals,articles, internal records, government publications etc
    * Economical and takes short time

### Types of data

* Structured data: row/columns -> tabular format -> excel file,.csv,.tsv etc
* Unstructured data: pictures,videos,text,.ppt,.pdf,log files ,voices
* Semi structured data: xml,json,html files {key:value}

### Data types in statistics

* Numerical data: data in the form of numbers
    *1.Discrete data: 
        * is a count that involves only integers. can't be subdivided into parts
        * ex: no of students in a class
        * no of workers in a company etc
        
    *2.Continuous data:
        * Continuous data could be meaniningfully divided into finger levels
        * It can be measured on a scale & have a numerical value
        * ex: height,weight of a children etc..
        
* Categorical data: made of words
* 1.Nominal data
                * nominal data is used for labelling variables
                * Ex: Gender(male,female)
* 2.Ordinal data
                * Ordinal data is data which is placed into some kind of order by their position on a scale
                * Ex: 1st ,2nd,3rd persons in a competition

```python
def grt(a,b):
    if a>b:
        return a
    else:
        return b
    
grt(90,87)
```

```python
grt([10,29,78],[90,65,76])
```

```python
vgrt=np.vectorize(grt)
vgrt([10,29,78],[90,65,76])
```
